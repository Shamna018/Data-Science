```python
#Programs to handle data using pandas
import pandas as pd
a=pd.read_csv('data.csv')
print(a)

#empty cell(NaN)
x=a['Calories'].mean()
a['Calories'].fillna(x,inplace=True)
print("\n\n Null values removed\n\n",a)

#date format
a['Date']=pd.to_datetime(a['Date'])
print("\n\nDate formate is corrected\n\n",a)

#Wromg duration
a.loc[3,'Duration']=45
print("\n\n Wrong duration data corrected\n\n",a)

#Duplicate data
b=a.duplicated().any()
print("\n\n Duplicated Datas are found.Duplication = ",b)
c=a.drop_duplicates()
print("\n\n Duplicate data are removed \n\n",c)
```
