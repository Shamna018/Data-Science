```python
#scatter graph
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,6,8])
y=np.array([3,5,2,8])
plt.scatter(x,y)
plt.show()
```

```python
#bar graph
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4])
y=np.array([1,2,6,8])
plt.bar(x,y)
plt.show()
```

```python
plt.barh(x,y)  #horizontal bar graph
plt.show()
```

```python
plt.bar(x,y,width=1) #width
plt.show()
```

```python
#histogram
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,6,8])
y=np.array([1,2,6,8])
plt.hist(x,y,width=2)
plt.show()
```

```python
#pie chart
import numpy as np
import matplotlib.pyplot as plt
x=np.array([110,90,60,80])
a=["apple","orange","mango","grapes"]
plt.pie(x,labels=a,shadow=True)
plt.show()
```

```python
y=([0,0.4,0,0])
plt.pie(x,labels=a,shadow=True,explode=y)  #explode
plt.show()
```

```python
#pie chart
import numpy as np
import matplotlib.pyplot as plt
x=np.array([50,50,60])
a=["apple","orange","mango"]
c=["lightblue","pink","lightgreen"]
plt.pie(x,labels=a,shadow=True,colors=c)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,6,8])
y=np.array([3,5,2,8])
m=np.array([1,2,3,4])
n=np.array([2,4,1,8])
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(m,n)
plt.show()
```
